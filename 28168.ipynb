```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
#อ่านข้อมูลจากไฟล์ study.csv
df =pd.read_csv('study.csv')
df.head()
```

```python
#ตรวจสอบว่ามีคอลัมอะไรบ้าง
df.columns
```

```python
input_columns =['Hours']#แก้
output_columns =['Scores']#แก้
```

```python
X=df.loc[:,input_columns]
y=df.loc[:,output_columns]
```

```python
#x.shape
```

```python
#splitting data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y ,test_size = .20,random_state =1,)
```

```python
#linear Regression
from sklearn.linear_model import LinearRegression
model1 = LinearRegression() #สร้าง model y=mx+b
model1.fit(X_train,y_train) #เรียกskค่า m กับ b
```

```python
#แสดง m b
print(model1.intercept_)#bจุดตัดแกน y
print(model1.coef_)#mความชัน
```

```python
#ทำนายบนชุดทดสอบ
y_p =model1.predict(X_test)
```

```python
#วัดผล MSE
from sklearn import metrics
print(np.sqrt(metrics.mean_squared_error(y_test,y_p)))
```

```python
#หากเรียน 7.5,8.5,9.25
X_quiz = [[7.5],[8.5],[9.25]]
y_quiz = model1.predict(X_quiz)
print(y_quiz)
```

```python
#ใช้ model ที่ได้หากเรียน 12.2 ,15.3 ,1 ,0 ,6 ,7.2
X_quiz =[[12.2],[15.3],[1],[0],[6],[7.2]]
y_quiz =  model1.predict(X_quiz)
print(y_quiz)
```

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
#อ่าข้อมูลจากไฟล์
df = pd.read_csv('insurance.csv')
df.head()
```

```python
#ตรวจคอลัม
df.columns
```

```python
#กำจัดค่าว่าง
df.dropna(inplace = True)
```

```python
input_columns = ['age','bmi','children']
output_columns= ['charges']
```

```python
X=df.loc[:,input_columns]
y=df.loc[:,output_columns]
```

```python
#splitting data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y ,test_size = .20,random_state =1,)
```

```python
#linear Regression
from sklearn.linear_model import LinearRegression
model1 = LinearRegression() #สร้าง model y=mx+b
model1.fit(X_train,y_train) #เรียกskค่า m กับ b
```

```python
#แสดง m b
print(model1.intercept_)#bจุดตัดแกน y
print(model1.coef_)#mความชัน
```

```python
#ทำนายบนชุดทดสอบ
y_p =model1.predict(X_test)
```

```python
#วัดผล MSE
from sklearn import metrics
print(np.sqrt(metrics.mean_squared_error(y_test,y_p)))
```

```python
#model ที่ได้ ทำนายเบี้ยประกัน
#age = 30 bmi =35 childen =1
#age = 19 bmi =25 childen =0
#age = 41 bmi =23 childen =2
X_quiz =[[30,35,1],[19,25,0],[41,23,2]]
y_quiz = model1.predict(X_quiz)
print(y_quiz)
```
